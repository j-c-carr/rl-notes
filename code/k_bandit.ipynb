# k-Armed Bandit Problems
This notebook goes over some of the problem solving approaches covered in Chapter 2 of Reinforcement Learning (Sutton and Barto)

```python
import numpy as np
from scipy import stats
from scipy.signal import lfilter
from scipy.special import softmax
import math as m
from random import randint

import matplotlib.pyplot as plt
%matplotlib inline 
```

### TestBed class

**Notation** (from the book)
* Q_star := true values of the actions
* Q := estimated values (list)
* N := values of the Nth reward 
* alpha := constant step size parameter, 0 < alpha <= 1
* c := confidence, used to control degree of exploration in UCB
* H := numerical preference

The TestBed has a few functions for the problem solving approaches discussed in Sutton Ch. 2.
- **Sample Average Method** (greedy / epsilon greedy)
- **Optimistic Initial Values**
- **Non stationary Action Values**
- **Upper Confidence Bound**
- **Gradient Bandit Alogrithm**

```python
class TestBed:
    def __init__(self, k, eps, n_timesteps, sigma_bandit = 1, mu_bandit=0):
        
        self.k = k
        self.eps = eps
        self.n_timesteps = n_timesteps
        self.R_avg = np.empty(n_timesteps)
        self.sigma_bandit = sigma_bandit
        self.mu_bandit = mu_bandit
        
        self.alpha = 1e-1
        self.c = 2
        
        self.init_estimate = 0
        
        self.method = {'sample_average': self.sample_average,
                       'non_stationary_action_value': self.non_stationary_action_value,
                      'upper_confidence_bound': self.upper_confidence_bound,
                      'gradient_bandit': self.gradient_bandit}
        
    def bandit(self, i, Q_star):
        """
        Rewards the agent for choosing the i-th bandit
        """
        return np.random.normal(Q_star[i], self.sigma_bandit)

    def set_initial_estimate(self, x):
        # Optimistic Initial Values
        self.init_estimate = x
    
    def sample_average(self):
        """
        1 run of sample average method
        """   
        Q_star = np.random.normal(self.mu_bandit, self.sigma_bandit, self.k)
        Q = np.empty(self.k)
        Q.fill(self.init_estimate) # Set initial bias
        N = np.zeros(self.k)
        
        # Track actions and rewards for each timestep
        R = np.empty(self.n_timesteps)
        AA = np.empty(self.n_timesteps)
        
        # Infinite loop
        for t in range(self.n_timesteps):
            p = np.random.binomial(1, 1-self.eps)
            A = p*np.argmax(Q) + (1-p)*randint(0, self.k-1)
            R_ = self.bandit(A, Q_star)
            N[A] += 1
            Q[A] += (R_- Q[A]) / N[A]
            
            R[t] = R_
            AA[t] = A

        return R
    
    def update_action_value(self, Q_star, sigma):
        """
        Changes the action values for non-stationary problem
        by taking a random walk of mean zero and variance sigma
        """
        for i in range(Q_star.shape[0]):
            Q_star[i] += np.random.normal(0, sigma)
        
        return
        
    def non_stationary_action_value(self, sigma_walk=0.1):
        """
        1 run of action value method with non stationary action values
        Using constant step size alpha
        """
        Q_star = np.random.normal(self.mu_bandit, self.sigma_bandit, self.k)
        Q = np.empty(self.k)
        Q.fill(self.init_estimate) # Set initial bias
        N = np.zeros(self.k)
        
        # Track actions and rewards for each timestep
        R = np.empty(self.n_timesteps)
        AA = np.empty(self.n_timesteps)
        
        # Infinite loop
        for t in range(self.n_timesteps):
            
            self.update_action_value(Q_star, sigma_walk)
            
            p = np.random.binomial(1, 1-self.eps)
            A = p*np.argmax(Q) + (1-p)*randint(0, self.k-1)
            R_ = self.bandit(A, Q_star)
            N[A] += 1
            Q[A] += self.alpha * (R_- Q[A])
            
            R[t] = R_
            AA[t] = A

        return R
    
    def upper_confidence_bound(self):
        """
        1 run of UCB, given confidence level c to
        control the degree of exploration
        """
        Q_star = np.random.normal(self.mu_bandit, self.sigma_bandit, self.k)
        Q = np.zeros(self.k)
        N = np.full(self.k, 1e-3) # To avoid dividing by zero
        
        # Track actions and rewards for each timestep
        R = np.empty(self.n_timesteps)
        AA = np.empty(self.n_timesteps)
        
        
        # infinite loop
        for t in range(self.n_timesteps):
            # TODO: optimize C (disgraceful!)
            
            # C := confidence term
            C = np.array([self.c*m.sqrt(m.log(max(1,t))) / N[i] for i in range(self.k) ])
            A = np.argmax(Q + C)
            R_ = self.bandit(A, Q_star)
            
            if(N[A] < 1):
                N[A] = 1
            else:
                N[A] += 1
            Q[A] += self.alpha * (R_- Q[A])
            
            R[t] = R_
            AA[t] = A

        return R
    
    def gradient_bandit(self):    
        """
        1 run of gradient bandit algorithm,
        """   
        Q_star = np.random.normal(self.mu_bandit, self.sigma_bandit, self.k)
        H = np.zeros(self.k)
        R_bar = 0
       
        
        # Track actions and rewards for each timestep
        R = np.empty(self.n_timesteps)
        AA = np.empty(self.n_timesteps)
        
        # Infinite loop
        for t in range(self.n_timesteps):
            # Choose an action with largest preference
            A = np.argmax(H)
            R_ = self.bandit(A, Q_star)
            
            # Update preferences
            H[A] += self.alpha * (R_ - R_bar) * (1 - softmax(H)[A])
            for i in range(H.shape[0]):
                if i != A:
                    H[i] -= self.alpha * (R_ - R_bar) * softmax(H)[A]
                    
            # Update average reward (assume nonstationary)
            R_bar += (1/max(1,t)) * (R_ - R_bar)
            
            R[t] = R_
            AA[t] = A

        return R
    def test(self, choice, n_runs=1):
        """
        Performes n_runs of the chosen algorithm,
        and computes the average for each timestep over N runs
        """
        R_tot = np.empty((n_runs, self.n_timesteps))
        for i in range(n_runs):
            R_tot[i] = self.method[choice]()

        self.R_avg = R_tot.mean(axis=0)

    def plot(self, label_str):
        """
        Filters data and plots R_avg after test()
        """
        n = 80  # the larger n is, the smoother curve will be
        b = [1.0 / n] * n
        a = 1
        smooth_R = lfilter(b,a,self.R_avg)
        plt.plot([i for i in range(self.n_timesteps)], smooth_R, 
                 linewidth=2, linestyle="-", label=label_str)

```

### Testing
- Larger reward variance

```python
Test1 = TestBed(10, .1, 1000, mu_bandit=4)
Test1.alpha = 1e-1
Test1.test("gradient_bandit", 10)
Test1.plot("gradient bandit")

#Test1.test("non_stationary_action_value", 10)
#Test1.plot("alpha = 1e-3")

Test1.test("upper_confidence_bound", 10)
Test1.plot("UCB")



plt.legend()
```
